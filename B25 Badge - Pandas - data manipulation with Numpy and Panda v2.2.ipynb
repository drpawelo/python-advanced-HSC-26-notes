### Modifying DataFrames

````python
# Change values: the old ('deprecated') way

data = pd.DataFrame({'names': ['Judy', 'Kim', 'Shaz', 'Sun'],
                     'study_year': [1,1,2,1], 
                     'avg_grades': [72,61,68,65] })

data.names[0] = "Natasha" # note that this way is likely to throw an error below
print(data)
````

````python
# Instead you would use the new .at or .loc operator. They do the same thing, but loc is more flexible

data = pd.DataFrame({'names': ['Judy', 'Kim', 'Shaz', 'Sun'],
                     'study_year': [1,1,2,1], 
                     'avg_grades': [72,61,68,65] })

data.loc[0, 'names'] = "Natasha" # Notice the order of arguments
data.loc[2, 'avg_grades'] = 100
print(data)
````

### **Side note about print vs implied return:** in Python notebooks there are two ways to 'show' what's going on with code

**print()** - 'show' many things, by 'printing/outputting' them below the cell. 

- You can print many things, and they will be shown in order you printed them (exaclty like with prinitng on paper)

**'implied output/return'** - You know return from functions. One thing can get returned from each function, and then the function 'ends/terminates'. - Only one thing is in an 'Out[]' block. It is the value of the last line in your code. 
- Some functions do not return anything (eg. print(...) does not). Also assignments = do not return anything

- ```by``` is a List of columns to sort the data by, in order. Items are sorted by first item in the List. If there is a tie, they are sorted by the second item, etc.
- ```inplace``` is a True/False value indicating whether the new value should be returned, or put back into the sorted dataframe. Use ```inplace=False``` if you want to print or output data, and use ```inplace=True``` if you want to change your actual data.
- ```ascending``` takes either True/False value, or a list of True/False values (if sorting by many columns)

### Mapping column values with .map( ) 

Previously when we used map, it was a Python method, into which we had to pass the List we wanted to map.

```map(mapping_function, my_list)```

That was a bit confusing, becuase methods usually are called on objects. It would make more sense for map to work like this:

```my_list.map(mapping_function)```

And Pandas give us the ability to do exactly that! Well... on Arrays, not Lists, but that's close enough.

```my_data_frame[column_name].map(mapping_function)```

**YOU CAN CHAIN .MAP( )** which makes certain tasks much simpler, like in 

```name.map(mapping_function_1).map(mapping_function_2)```


````python
data = pd.DataFrame({'fruits' : ["banana",  "kiwi", "apple"]})

data['lengths'] = data['fruits'].map(len)

data
````

### But first... a short catchup on lambda functions

````python
# Traditional function definition looks like this:
data = pd.DataFrame({'fruits' : ["banana",  "kiwi", "apple"]})

def first_letter(word):
    return word[0]

print(first_letter("banana"))


# And then you can use this function's NAME in a map

data['first_letters'] = data['fruits'].map(first_letter)

data
````

Quick recap about lambda functions:

````python
# 'lambda function' is Python's simplified syntax for defining functions.
# They do not really have a name, but rather you can put the function definition right into your code
# Function definition looks like this: 
# lambda inputs: outputs

data = pd.DataFrame({'fruits' : ["banana",  "kiwi", "apple"]})

first_letter = lambda word: word[0] # This replaces function definition. But is a bit harder to read.
# You sort of put a function in a variable. It can still be used like a function:
print(first_letter("banana"))


data['first_letters'] = data['fruits'].map(first_letter)

data
````

````python
# But most commonly lambdas are used in situations where calculation is quick and simple
# And there is no need to give it a special name, or to reuse it later 

data = pd.DataFrame({'fruits' : ["banana",  "kiwi", "apple"]})

data['first_letters'] = data['fruits'].map(lambda word: word[0])

data
# this is THE MOST COMMON SCENARIO
````

````python
# Another example
offices = pd.DataFrame({'department': ['sales'] * 3 + ['marketing'] * 5 + ['r&d'] * 4,
                        'floor': [1,2,3,1,2,3,3,3,1,1,1,2]})
offices['floor_sign'] = offices['floor'].map(lambda floor: f"Floor {floor}")
offices
````

````python
offices = pd.DataFrame({'department': ['sales'] * 3 + ['marketing']*5 + ['r&d'] * 4, 
                        'floor': [1,2,3,1,2,3,3,3,1,1,1,2]})

floor_names = {1: 'Ground Floor', 2: 'Main Floor', 3: "Roof Floor"}

offices['floor_sign'] = offices['floor'].map(lambda floor: floor_names[floor] )
offices
````

### Cleaning up data with .map( )

Here's an **example of chaining** ```.map( )``` to clean up data:

````python
# Data is so messy! Read it. Do you see any patterns? How could it be fixed?

offices = pd.DataFrame({
    'department': ['-Sales-', 'sales  ', '-SALES-', 'marketing', 'MARKETING', '-marketing-',
                   'Marketing', '  marketing', '-R&D-', ' r&d', 'r&d  ', 'r&d'],
    'floor':[1,2,3,1,2,3,3,3,1,1,1,2]
})
offices
````

Look at this data. What is wrong with 'department'? and how would you fix it?

Let's clean up capitalisation and such, and then replace department names with unified 'codes', so that we never have this problem again. So that  `marketing`, `MARKETING`,`-marketing-` and `Marketing` all become "MAR".

````python
# let's first start with just one string
ugly_data_point = "-MARKETING-"
department_codes = {'sales': "SAL", 'marketing': "MAR", 'r&d': "RND" }

# Let's create lambdas which will clean it up, one source of messiness at a time:
remove_space_and_dash = lambda word: word.strip().strip('-')
to_lower_case = lambda word: word.lower()
dept_name_to_code = lambda dept_name: department_codes[dept_name]

# for this one datapoint we can call each function in turn:
print(ugly_data_point)
ugly_data_point = remove_space_and_dash(ugly_data_point)
print(ugly_data_point)
ugly_data_point = to_lower_case(ugly_data_point)
print(ugly_data_point)
ugly_data_point = dept_name_to_code(ugly_data_point)
print(ugly_data_point)
````

You could even chain use of those three functions, but it will look a bit confusing

Remember: code nested within brackets 'happens' from inside to outside, like in `(2+(3*4))/7` which produces `(2+(12))/7` then `(14)/7` then `2`

````python
ugly_data_point = "-MARKETING-"
ugly_data_point = dept_name_to_code(to_lower_case(remove_space_and_dash(ugly_data_point)))
print(ugly_data_point)
````

Now, in Pandas we would say: `map` all items in the column `department` with the result of putting them through `this function`. And then `that function`...

Remember: code daisychained with `.` happens from left to right e.g. `Aabc".strip("A").upper()` will produce `ABC`


````python
# Mapping - add new column, with cleaned up values
offices['dept_code'] = offices['department'].map(remove_space_and_dash).map(to_lower_case).map(dept_name_to_code)
offices
````

````python

# Mapping - add new column, with cleaned up values
offices['dept_code'] = offices['department'].map(remove_space_and_dash).map(to_lower_case).map(dept_name_to_code)
offices
````

### DataFrame and Higher Order Functions: .map( ) and  .apply( ) 

Both of these do something very similar, but in simplest terms:

- **map( )** works with column items, one at a time, or the whole DataFrame. Use it to represent each item as another item.
- **apply( )** access each row, or column, and represent it as one value (reduce it to one value) 

````python
# Here's a silly example:
# map() changes ONE column 

people = pd.DataFrame({
    'firstname': ['Jill', 'Ryu', 'Rashni'],
    'lastname':['McDoughal', 'Kawasaki', 'Ng']
})

to_upper_case = lambda word: word.upper()
people['firstname'] = people['firstname'].map(to_upper_case)
people
````

````python
# Here's another silly example:
# map() to change ALL columns 

people = pd.DataFrame({
    'firstname': ['Jill', 'Ryu', 'Rashni'],
    'lastname':['McDoughal', 'Kawasaki', 'Ng']
})

to_upper_case = lambda word: word.upper()
people = people.map(to_upper_case)
# This could also be written with lambda directly in the map (can get messy, if complex)
# people = people.map(lambda word: word.upper())

people
````

````python
# And another silly example: 
# apply() represent values as one value 
# note: ‘columns’ means apply function to each row 

people = pd.DataFrame({
    'firstname': ['Jill', 'Ryu', 'Rashni'],
    'lastname':['McDoughal', 'Kawasaki', 'Ng']
})

join_names = lambda column: f"{column['firstname']} {column['lastname']}"

people['fullname'] = people.apply(join_names, axis='columns')
people
````

### Calculate a new column from rows with ```.assign( )```

First, why do we need this? If you want to create a new column, you could just use the good ole Python notation like

```
my_data[column_name] = some calculation
```

Notice right hand side (which happens first) tells you what to do, and left side tells you where to assign it. 

But if you are creating many columns it can get long. But also often you would prefer to 'pipe' your data into the next thing, then next thing, then next thing.

Assign allows you to create many columns at one go. By just assigning values or performing calculations. It is more than just **syntax sugar** - it allows you to write long piped code, familiar from other languages for data analytics.

````python
# Assign new columns to a DataFrame. 
# Existing columns that are re-assigned will be overwritten.

foods = pd.DataFrame({
    'name':['Bagel', 'Milk Chocolate', 'Carrot', 'Ham Sandwich', 'Egg Cake'],
    'diet':['Vegan', 'Vegetarian', 'Vegan', 'Meat', 'Vegetarian'],
    'supplier':['Bros', 'Luca', 'Whitmore', 'Union', 'Lovecrumbs'],
    'price':[4.30, 2.10, 0.7, 5.70, 3.20]
})

# create two new columns: student_price and available (this could be in one line)
foods = foods.assign(
                     student_price = foods['price'] * 0.9, 
                     available = True
                     )
foods
````

Note about `inplace=True`. inplace is awesome, but sometimes you can't use it. This is one of those special moments.

For a change, **assign( ) does not change the original DataFrame**. That's because if you specified 'inplace=True' it would just add a column called 'inplace' and put values True in every row of that column. It's just a peculiar price we need to pay for the power of assign.

so you need to do 

`my_data = my_data.assign(new_column = some calculation)`

but these two below would not work:

`my_data.assign(new_column = some calculation)` or
`my_data.assign(new_column = some calculation, inplace = True)`

````python
foods = pd.DataFrame({
    'name': ['Bagel', 'Milk Chocolate', 'Carrot', 'Ham Sandwich', 'Egg Cake'],
    'diet':['Vegan', 'Vegetarian', 'Vegan', 'Meat', 'Vegetarian'],
    'supplier':['Bros', 'Luca', 'Whitmore', 'Union', 'Lovecrumbs'],
    'price':[4.30, 2.10, 0.7, 5.70, 3.20]
})

foods.assign(student_price = foods['price'] * 0.9, inplace=True)
# this will go wrong in two ways! Do you see one? Run next cell to see the other bug.
````

Above you will likely see a big red warning here. Keep reading to understand why it is, and what to do. Read the error (errors are your friends!), do you think you could fix it yourself?

Error says, instead of 

`df[col].method(value, inplace=True)`
try using 
`df.method({col: value}, inplace=True)` or 
`df[col] = df[col].method(value)`

So instead of

`foods['sold_since_year'].replace(0, np.nan, inplace=True)`

it would be safer to use

`foods['sold_since_year'] = foods['sold_since_year'].replace(0, np.nan)`

### Agg() or Aggregate() - aggregate and summarise in many ways at the same time!

aggregate() take a list of names of functions to use on your data. You can do it on the whole dataframe, or just on one column.

````python
# spot the difference, this does the same as previous way, but we specify column 'price' later
# both ways are ok
foods.groupby(foods['diet']).price.agg(['mean', 'max', 'median', 'sum'])
````

````python
# you could even write your own aggregation functions!
def window(column):
    return max(column) - min(column)

foods.groupby(foods['diet']).price.agg(['mean','median', 'sum', 'max', 'min',  window])
````

````python
# sure, you could do it with lambdas, but you miss a chance to call the column:
foods.groupby(foods['diet']).price.agg(['mean','median', 'sum', 'max', 'min', lambda col: max(col) - min(col)])
````

And the very useful part of **agg** is that you can use different aggregators for different columns. It will become  more readable too. To do this you would use good old Dictionary: keys are columns, values are lists with aggregators to use:

say to get a 'mean' and 'median' of column `price`, but 'sum', 'max', 'min' of column `sold_since_year` you would write:

```
.agg(
      {
        'price':['mean','median'],
        'sold_since_year': ['sum', 'max', 'min']
      }
    )
```
                                

````python
foods = pd.DataFrame({
    'name': ['Bagel', 'Milk Chocolate', 'Carrot', 'Ham Sandwich', 'Egg Cake', 'Tap Water'],
    'diet':['Vegan', 'Vegetarian', 'Vegan', 'Meat', 'Vegetarian', 'Vegan'],
    'supplier':['Bros', 'Luca', 'Whitmore', 'Union', 'Lovecrumbs', 'Water Tap'],
    'price':[4.30, 2.10, 0.7, 5.70, 3.20, 0],
    'sold_since_year': [2018, 2015, 0, 2012, 0, 0]
})

foods.groupby(foods['diet']).agg({'price':['mean','median'],
                                  'sold_since_year': ['sum', 'max', 'min', window]
                                })
````

# Categories - Grouping results by a range of values. Use ```pd.data.cut( data, bins, labels )``` 

Often we want to categorise our data into particular groups by value. Given a set of values, we want to decide in which range they belong.

Imagine a bunch of student exam scores (70, 54, 40, 66) that we want to translate into grades (A, B, C, D, F).

We will need a key of where one grade ends and another starts. One way to call them are bins (like buckets/containers) and our task is to put each score in one of these bins.

- F is (0, 40]
- D is (40, 50]
- C is (50, 60]
- B is (60, 70]
- A is (70, 100]

Note: 

- '(' means the value is included in the bin
- '[' means the value is excluded

In Pandas you could describe it as ```[(0, 40] < (40, 50] < (50, 60] < (60, 70] < (70, 100]]```

````python
import pandas as pd
import numpy as np

student_scores = [40, 42, 46, 54, 60, 63, 66, 70, 72]

bins = [0, 40, 50, 60, 70, 100]
grades = ["F", "D", "C", "B", "A"]

# Note: labels (eg. grades) are optional, but very useful

# cut will categorise
categories = pd.cut(student_scores, bins, labels=grades, right=False)
print(categories) # Print shows how are values are categorised, and also what do categories mean
````

````python
# If you want the rightmost elements to be included in the smaller category (eg. for score 40 to be an 'F')
# Use right=True argument, or just no right argument (True is a default)
categories = pd.cut(student_scores, bins, labels=grades, right=True)
print(categories)
print()
````

When you use ```pd.cut(student_scores, bins, labels=["F","D","C","B","A"])``` the resulting object contains information about 

- Which category each of your data points belongs to
- What are the categories and their boundaries

````python
student_scores = [40, 54, 60, 66, 70]
bins = [0, 40, 50, 60, 70, 100]
labels = ["F", "D", "C", "B", "A"]
categories = pd.cut(student_scores, bins, labels=labels)

print( categories )
print()
print( categories.tolist() )
print( categories.codes ) # in older versions this was called .labels
````

````python
pd.Series(categories).value_counts() 
````

````python
# There are times when cumulative sum is very useful. It will add all items until now, but the order is peculiar
pd.Series(categories).value_counts().cumsum() 
````
